## Operators in Python

Operators are special symbols that perform operations on variables and values. They are essential for performing calculations, comparisons, and logical operations in programs.


### Arithmetic Operators

Arithmetic operators perform mathematical calculations. They include addition (`+`), subtraction (`-`), multiplication (`*`), division (`/`), floor division (`//`), modulus (`%`), and exponentiation (`**`).


```python
## Arithmetic Operators
a = 10
b = 4

add_result = a+b
print("Result of a + b:", add_result)
sub_result = a-b
print("Result of a - b:", sub_result)
mul_result = a*b
print("Result of a * b:", mul_result)
div_result = a/b
print("Result of a / b:", div_result)
floor_div_result = a//b
print("Result of a // b:", floor_div_result)
mod_result = a%b
print("Result of a % b:", mod_result)
exp_result = a**b
print("Result of a ** b:", exp_result)
```

### Comparison Operators

Comparison operators compare two values and return a boolean (True or False). They are used in conditions, loops, and decision-making to control program flow.


```python
## Comparison Operators
x = 5
y = 10
print("x == y:", x == y)  # False
print("x != y:", x != y)  # True
print("x > y:", x > y)    # False
print("x < y:", x < y)    # True
print("x >= y:", x >= y)  # False
print("x <= y:", x <= y)  # True

name1= "AnubhavC"
name2= "Anubhav"
print("name1 != name2:", name1 != name2)
print("name1 < name2:", name1 < name2)
print("name1 >= name2:", name1 >= name2)
```

### Logical Operators

Logical operators combine boolean values and return a boolean result. They are used in conditions to make complex decisions. The three main logical operators are `and`, `or`, and `not`.

#### Truth Table for `and` Operator

| A     | B     | A and B |
| ----- | ----- | ------- |
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |

#### Truth Table for `or` Operator

| A     | B     | A or B |
| ----- | ----- | ------ |
| True  | True  | True   |
| True  | False | True   |
| False | True  | True   |
| False | False | False  |

#### Truth Table for `not` Operator

| A     | not A |
| ----- | ----- |
| True  | False |
| False | True  |


```python
## Logical Operators (and, or, not)
is_raining = False
is_windy = True

print("Is it raining and windy?", is_raining and is_windy)  # False
print("Is it raining or windy?", is_raining or is_windy)   # True
print("Is it not raining?", not is_raining)                # True
```
