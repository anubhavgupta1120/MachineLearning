## Variables in Python

Variables are containers for storing data values. This section teaches how to declare, assign, and print variables of different types (integers, floats, strings, booleans).


### Variable Declaration and Printing

This code demonstrates assigning values to variables of different data types and printing them using the `print()` function. Understanding data types is fundamental for writing effective Python programs.


```python
age=26
height=5.8
name="Anubhav"
is_student=True
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
```

### Type Conversion

Type conversion (also called type casting) allows you to change a variable's data type to another compatible type. This is useful when you need to perform operations that require specific types, like converting user input (strings) to numbers for calculations.


```python
age='25'
age = int(age)

print(type(age))

## Not allowed to convert string to integer
name="Anubhav"
## int (name)

height = 5.8
type(height)
height = int(height)
print(type(height))
print(height)
height = float(height)
print(type(height))
print(height)
```

### Dynamic Typing

Python is dynamically typed, meaning you can reassign a variable to a value of a different type without declaring the type explicitly. This flexibility makes Python easy to use but requires careful type management to avoid runtime errors.


```python
# Python allows the type of variable to change as the program executes. This is known as dynamic typing. It allows you to assign different types of values to the same variable without any issues.

var = 10
print(var, type(var))  # Output: 10 <class 'int'>
var = "Now I'm a string"
print(var, type(var))  # Output: Now I'm a string <class 'str'>
```

### User Input

The `input()` function allows your program to accept user input from the keyboard. It always returns a string, so you often need to convert it to other types (like `int()`) for numerical operations. This is essential for interactive programs.


```python
age=int(input("Enter your age: "))
print("You entered:", age, type(age))
```

### Simple Calculator Coding Exercise

This exercise combines variables, user input, type conversion, and basic arithmetic operations. It demonstrates how to build a simple interactive program that performs calculations based on user-provided numbers.


```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
add = num1+num2
subtract = num1-num2
multiply = num1*num2
divide = num1/num2
print("Addition:", add)
print("Subtraction:", subtract)
print("Multiplication:", multiply)
print("Division:", divide)
```
