## Case Sensitivity

Python is case-sensitive, which means `name` and `Name` are two different variables. This example shows how to assign string values to variables and print them.


```python
## Basic Syntax and Data Types in Python

name="Anubhav"
Name="Gupta"
print(name)
print(Name)
```

### Control Flow, Indentation, and Line Continuation

This cell demonstrates basic control flow using an `if`/`else` statement and why consistent indentation matters in Python (it defines code blocks).

It also shows:

- How comments (`# ...`) can explain what your code does without affecting execution.
- How to continue a long expression across lines using a backslash (`\\`).
- That multiple statements can appear on the same line using a semicolon (`;`), though this is not common in idiomatic Python.


```python
age=25
if age > 18:
  print("You are an adult.")
  print(age)
else:
  print("You are a minor.")
print("This is a single-line comment in Python.")

## Line Continuation in Python (Backslash \)
total = 1 + 2 + 3 + \
4+5
print(total)

#Multiple statements on a single line using semicolon ;
x = 10; y = 20; z = x + y
print(z)
```

### Types, and Semantics

This code demonstrates assigning values to variables and checking their types using `type()`.

It also highlights the difference between syntax (how code is written) and semantics (what the code means) in Python.


```python
## Understanding semantics and syntax in Python

## variable assignment
age=25 ## age is a number
name="Anubhav" ## name is a string
print(type(age))
print(type(name))
```

### Dynamic Typing & Type Inference

Python is dynamically typed, which means a variable can hold values of different types over time. This example shows how `x` can be an integer, then become a string.

Use `type()` to inspect the current type of a variable at runtime.


```python
## Type Inference in Python
x = 10
print(type(x))
x = "Hello"
print(type(x))
```

### Nested Blocks and Indentation

Python uses indentation to determine which statements belong together. Nested blocks (like an `if` inside another `if`) must be indented consistently.

This example shows how inner blocks are skipped when their condition is `False`.


```python
## Code examples of indentation in Python
if True:
  print("This is indented code block.")
  if False:
    print("This will not be printed.")
  print("This is outside the inner if block.")
print("This is outside the if block.")
```
